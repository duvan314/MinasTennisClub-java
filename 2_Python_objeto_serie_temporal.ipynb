Importing the Packages

```python
import pandas as pd
import numpy as np
```

```python
raw_csv_data = pd.read_csv("Index2018.csv")
df_comp = raw_csv_data.copy()
```

From Text to Date - convertir de texto a fecha



```python
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
```

```python
df_comp.head()
```

```python
df_comp.date.describe()
```

Setting the index

```python
df_comp.set_index("date", inplace = True)
```

```python
df_comp.head()
```

```python
# df_comp.date.describe() #esperar obtener un mensaje de error porque ya no tenemos una columna "fecha" ya que es el índice
```

Setting the Desired Frequency

h: houurly (horaria)
w: weekly (semanal)
d: daily (diaria)
m: monthly (mensualmente)
a: anual

```python
df_comp = df_comp.asfreq('d') # d: diaria
```

```python
df_comp.head()
```

b:business days (días laborales)

```python
df_comp = df_comp.asfreq('b') ## solo considera los dias laborales
```

```python
df_comp.head()
```

Handlind Missing Values

```python
df_comp.isna()
```

```python
df_comp.isna().sum()
```

```python
df_comp
```

```python
df_comp.spx = df_comp.spx.fillna(method = 'ffill') ## ffill = front filling dia posterior 
```

```python
df_comp.isna().sum()
```

```python
df_comp.ftse = df_comp.ftse.fillna(method = 'bfill') ## back filing dia anterior 
```

```python
df_comp.dax = df_comp.dax.fillna(value = df_comp.dax.mean()) ## no se suele usar para datos de series temporales
```

```python
df_comp.nikkei = df_comp.nikkei.fillna(method = 'bfill') ## back filing dia anterior 
```

Simplifyind the Dataset

```python
df_comp['market_value'] = df_comp.spx ## se crea la columna 'market_value'
```

```python
df_comp.describe()
```

```python
del df_comp['spx'] ## eliminar una sola columna
```

```python
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
```
